## Benchmark Subset Metadata

```python
import json

with open('/content/drive/MyDrive/Orfium/da-tacos_metadata/da-tacos_benchmark_subset_metadata.json') as f:
	benchmark_metadata = json.load(f)

# Initialize a counter for performance IDs
total_p_ids = 0

# Loop through all W_ids and count P_ids
for work_id in benchmark_metadata:
    total_p_ids += len(benchmark_metadata[work_id])

# Print the total number of P_ids
print(f"Total number of W_ids: {len(benchmark_metadata.keys())}")
print(f"Total number of P_ids: {total_p_ids}")
```

```python
import random

# Get a list of all keys from the metadata
keys = list(benchmark_metadata.keys())

# Randomly select 5 keys within the range from index 10 to 15
selected_keys = random.sample(keys[0:100], 3)

# Pretty print the selected elements
for key in selected_keys:
    print(f"Key: {key}")
    print(json.dumps(benchmark_metadata[key], indent=4))
    print("-" * 80)
```

```python
import pandas as pd

# Initialize an empty list to store the data
data = []

# Loop through the JSON data (assuming `benchmark_metadata` is your JSON object)
for work_id, performances in benchmark_metadata.items():
    for perf_id, metadata in performances.items():
        # Collect relevant data into a dictionary
        row = {
            'work_id': work_id,
            'performance_id': perf_id,
            'work_title': metadata.get('work_title', None),
            'work_artist': metadata.get('work_artist', None),
            'perf_title': metadata.get('perf_title', None),
            'perf_artist': metadata.get('perf_artist', None),
            'release_year': metadata.get('release_year', None),
            'instrumental': metadata.get('instrumental', None),
            'perf_artist_mbid': metadata.get('perf_artist_mbid', None),
            'mb_performances': metadata.get('mb_performances', None)
        }
        data.append(row)

# Convert the list of dictionaries into a pandas DataFrame
benchmark_df = pd.DataFrame(data)

# Display the DataFrame
print(benchmark_df.shape)
display(benchmark_df.sample(10))
```

## Cover Analysis Metadata


```python
with open('/content/drive/MyDrive/Orfium/da-tacos_metadata/da-tacos_coveranalysis_subset_metadata.json') as f:
	coveranalysis_metadata = json.load(f)

# Initialize a counter for performance IDs
total_p_ids = 0

# Loop through all W_ids and count P_ids
for work_id in coveranalysis_metadata:
    total_p_ids += len(coveranalysis_metadata[work_id])

# Print the total number of P_ids
print(f"Total number of W_ids: {len(coveranalysis_metadata.keys())}")
print(f"Total number of P_ids: {total_p_ids}")
```

```python
# Get a list of all keys from the metadata
keys = list(coveranalysis_metadata.keys())

# Randomly select 5 keys within the range from index 10 to 15
selected_keys = random.sample(keys[0:100], 3)

# Pretty print the selected elements
for key in selected_keys:
    print(f"Key: {key}")
    print(json.dumps(coveranalysis_metadata[key], indent=4))
    print("-" * 80)
```

```python
# Loop through the JSON data
for work_id, performances in coveranalysis_metadata.items():
    for perf_id, metadata in performances.items():
        # Extract the basic fields
        row = {
            'work_id': work_id,
            'performance_id': perf_id,
            'work_title': metadata.get('work_title', None),
            'work_artist': metadata.get('work_artist', None),
            'perf_title': metadata.get('perf_title', None),
            'perf_artist': metadata.get('perf_artist', None),
            'release_year': metadata.get('release_year', None),
            'instrumental': metadata.get('instrumental', None),
            'perf_artist_mbid': metadata.get('perf_artist_mbid', None),
        }

        # Extract the performance lengths from the mb_performances field
        for performance_id, performance_data in metadata.get('mb_performances', {}).items():
            performance_row = row.copy()
            performance_row['performance_id'] = performance_id  # Update performance ID for each entry
            performance_row['performance_length'] = performance_data.get('length', None)
            data.append(performance_row)

# Convert the list of dictionaries into a pandas DataFrame
coveranalysis_df = pd.DataFrame(data)

# Display the DataFrame
print(coveranalysis_df.shape)
display(coveranalysis_df.sample(10))
```
